```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
file_path = '/content/drive/My Drive/Prodigy_Infotech/titanic.csv'
data = pd.read_csv(file_path)
print(data)
```

```python
missing_values=data.isnull().sum()
print(missing_values)
```

```python
mean_age=data['Age'].mean()
data["Age"].fillna(mean_age,inplace=True)
print(data)
```

```python
missing_values=data.isnull().sum()
print(missing_values)
```

```python
data['Cabin'].fillna('R23', inplace=True)
print(data)
```

```python
mode_Embarked=data['Embarked'].mode()[0]
data["Embarked"].fillna(mode_Embarked,inplace=True)
print(data)
```

```python
missing_values=data.isnull().sum()
print(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Survival Counts**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=data)
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.title('Survival Counts')
plt.show()
```

# **Survival Counts By Sex**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survival Counts by Gender')
plt.show()
```

# **Survival Counts by Pclass**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Counts by Passenger Class')
plt.show()
```

# **Relationship between Age and Survival**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=data)
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.title('Age vs Survival')
plt.show()
```

# **Survival Rates By Embarked**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survival Counts by Embarked')
plt.show()
```

```python
survival_counts = data['Survived'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(survival_counts, labels=['Not Survived', 'Survived'],autopct='%1.1f%%', startangle=90,colors=['Skyblue', 'pink'])
plt.title('Survival Distribution on the Titanic')
plt.axis('equal')
plt.show()
```
